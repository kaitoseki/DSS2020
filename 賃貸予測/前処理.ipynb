## バス・トイレ
- 専用バス/専用トイレ/浴室乾燥機/温水洗浄便座/追焚機能/洗面台独立

```python
train['バス・トイレ'].value_counts().head(10)
```

```python
train_bath = train["バス・トイレ"].str.split("\t", expand=True)
train_bath
```

```python
bath_calums = ['専用バス', '専用バス／', '専用トイレ', '専用トイレ／','バス・トイレ別','バス・トイレ別／','シャワー','シャワー／',
              '追焚機能', '追焚機能／', '浴室乾燥機', '浴室乾燥機／','温水洗浄便座','温水洗浄便座／','洗面台独立','洗面台独立／']
```

```python
# カルムの数
count=len(train_bath.columns)

# bath_calumsの追加
for _, colname in enumerate(bath_calums):
    train_bath[colname]=0
        
    for k in range(count):
        train_bath.loc[train_bath[k]==colname, colname]=1 
        
# 必要なカルムのみ        
train_bath = train_bath[bath_calums].copy()
```

```python
#train_bath[]=train_bath[]+train_bath[]
train_bath['専用バス']=train_bath['専用バス']+train_bath['専用バス／']
train_bath['専用トイレ']=train_bath['専用トイレ']+train_bath['専用トイレ／']
train_bath['バス・トイレ別']=train_bath['バス・トイレ別']+train_bath['バス・トイレ別／']
train_bath['シャワー']=train_bath['シャワー']+train_bath['シャワー／']
train_bath['追焚機能']=train_bath['追焚機能']+train_bath['追焚機能／']
train_bath['浴室乾燥機']=train_bath['浴室乾燥機']+train_bath['浴室乾燥機／']
train_bath['温水洗浄便座']=train_bath['温水洗浄便座']+train_bath['温水洗浄便座／']
train_bath['洗面台独立']=train_bath['洗面台独立']+train_bath['洗面台独立／']
```

```python
train_bath.drop(['専用バス／', '専用トイレ／', 'バス・トイレ別／', 'シャワー／', '追焚機能／', '浴室乾燥機／', '温水洗浄便座／',
                 '洗面台独立／'], axis=1)
```
